```python
from keras.models import Sequential
from keras.layers import Dense
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('train.csv')
data.head()
```

```python
X = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

```python
X.shape
```

```python
y.value_counts()
```

```python
model = Sequential()
model.add(Dense(units = 20,input_dim = 20, activation='relu'))
model.add(Dense(units = 20, activation = 'relu'))
model.add(Dense(units = 10, activation= 'relu'))
model.add(Dense(units = 4, activation= 'softmax'))
```

```python
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(X,y,epochs=800, batch_size=50)
```

```python
loss, accuracy = model.evaluate(X,y)
print(f'Loss:{loss}, Accuracy:{accuracy}')
```

```python
predict = model.predict(X)
print(*predict, sep=' ')
```

```python
train = pd.read_csv('test.csv')
test = train.drop('id',axis=1)
```

```python
predict_y = model.predict(test)
```

```python
n=[]
for i in predict_y:
  m =[]
  for j in i:
    m.append(np.round(j))
  n.append(m)
```

```python
k=[]
for i in n:
  for j in i:
    if j==1:
      k.append(i.index(j))
```

```python
train['Predictions'] = k
```

```python
train.head()
```

```python
train.to_csv('Predictions.csv')
```
